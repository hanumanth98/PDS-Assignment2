**1b. Removing the text from float values**

```python
dframe['Mileage'] = dframe['Mileage'].apply(lambda x: float(str(x).split()[0]))
dframe['Engine'] = dframe['Engine'].apply(lambda x: float(str(x).split()[0]))
dframe['Power'] = dframe['Power'].apply(lambda x: float(str(x).split()[0]))
dframe['New_Price'] = dframe['New_Price'].apply(lambda x: float(str(x).split()[0]))

column_types = dframe.dtypes
column_types
```

**1a. Search for any missing values, select a better option, and replace them**

```python
dframe.hist(figsize=(10, 8))
plt.show()
```

```python
dframe['Seats'].value_counts()
```

```python
dframe.Mileage.fillna(dframe.Mileage.mean(), inplace=True)
dframe.Engine.fillna(dframe.Engine.median(), inplace=True)
dframe.Power.fillna(dframe.Power.median(), inplace=True)
dframe.Seats.fillna(dframe.Seats.mode()[0], inplace=True)
dframe = dframe.drop('New_Price', axis = 1)
```

**Explanation**

1. Since the mileage column has a normal distribution, missing values are replaced with the mean.
2. Because the distributions of the Power and Engine columns were skewed, the median was chosen.
3. The numbers in the seats column are dominated by one value: more cars had five seats. utilized mode to substitute missing values as a result.
4. The new price column is unsuitable for imputing since it contains more empty values. therefore I removed the column.

1c. Transform text for categorical variables into a single hot encoding value.

```python
one_hot_encoded = pd.get_dummies(dframe['Fuel_Type'])
dframe = pd.concat([dframe, one_hot_encoded], axis=1)
dframe.drop('Fuel_Type', axis=1, inplace=True)

one_hot_encoded = pd.get_dummies(dframe['Transmission'])
dframe = pd.concat([dframe, one_hot_encoded], axis=1)
dframe.drop('Transmission', axis=1, inplace=True)
```

1d. Create one more feature.

```python
import datetime

dframe['Kms_per_Age'] = dframe['Kilometers_Driven']

for i in range(len(dframe['Year'].tolist())):
    current_year = datetime.datetime.now().year
    age = current_year - dframe.loc[i, 'Year']
    if age != 0:
        kms_per_age = dframe.loc[i, 'Kilometers_Driven'] / age
    else:
        kms_per_age = dframe.loc[i, 'Kilometers_Driven']

    dframe.loc[i, 'Kms_per_Age'] = kms_per_age

dframe
```
